## Basic Operations on Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def pltImage(img):
    plt.imshow(img)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
img = cv2.imread('image1.png')
px = img[100,100]
print(px)
```

```python
pltImage(img)
```
